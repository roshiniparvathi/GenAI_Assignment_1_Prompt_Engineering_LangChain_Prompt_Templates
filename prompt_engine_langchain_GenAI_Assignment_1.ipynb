## Task 1: Replace Hardcoded Prompts

```python
# Create a reusable prompt template
template = "Explain {topic} in simple terms for beginners"

# Initialize PromptTemplate
prompt = PromptTemplate(input_variables=["topic"], template=template)

# Generate output
print("Output:", prompt.format(topic="Artificial Intelligence"))
```

## Task 2: Multi-Input Prompt System

```python
# Create multi-input template
template = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test cases
print("Test 1:", multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print("Test 2:", multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print("Test 3:", multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine

```python
# Define different styles
teaching_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} clearly step by step")
interview_template = PromptTemplate(input_variables=["topic"], template="Ask 3 questions about {topic}")
story_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} as a story")

topic = "Machine Learning"

print("Teaching:", teaching_template.format(topic=topic))
print("Interview:", interview_template.format(topic=topic))
print("Story:", story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

```python
# Create chat template with roles
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

# Generate messages
messages = chat_template.format_messages(role="teacher", topic="Neural Networks")

for msg in messages:
    print(msg)
```

## Task 5: Input Validation Layer

```python
# Validate inputs
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to beginner.")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to formal.")
        tone = "formal"

    return audience, tone
```

## Task 6: Prompt Generator App

```python
# Generate dynamic prompt
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
    }

    if style not in styles:
        print("Invalid style! Defaulting to teaching.")
        style = "teaching"

    template = styles[style]

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)

# Example outputs
print("Example:", generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print("Extra Test:", generate_prompt("AI", "expert", "formal", "teaching"))
```

## Task 7: Template Reusability Test

```python
# Reuse single template
reuse_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} in simple terms")

topics = ["AI", "Python", "Blockchain", "Cloud Computing", "Data Science"]

for t in topics:
    print("Output:", reuse_template.format(topic=t))
```

## Final Pipeline Flow

User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output
